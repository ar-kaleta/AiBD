```python
Xcopy = X.copy()
```

```python
mapping = {"Not at all worried" : "Not worried", "Not so worried" : "Not worried"}
```

```python
Xcopy["In general, how worried are you about earthquakes?"] = \
Xcopy["In general, how worried are you about earthquakes?"].apply(lambda row: mapping.get(row, "Worried"))
Xcopy
```

```python
ct = pd.crosstab(Xcopy["US Region"], Xcopy["In general, how worried are you about earthquakes?"])
ct
```

```python
ct.plot.bar(figsize=(15,7))
plt.tight_layout()
```

```python
(ct["Not worried"]/ct["Worried"])
```

```python
(ct["Not worried"]/ct["Worried"]).plot.bar(figsize=(15,7))
plt.tight_layout()
plt.title('Stosunek osób nie martwiących się o trzęsienie ziemi do osób obawiających się trzęsienia ziemi')
```

```python
(ct["Not worried"]/ct["Worried"]).idxmax(), (ct["Not worried"]/ct["Worried"]).idxmin()
```

Stosunek osób które nie boją się trzęsienia ziemi do stosunku osób które boją się trzęsienia ziemi jest największy w regionine East North Central, natomiast największy w regionie Pacific. Można zatem powiedzieć, że najbardziej trzęsienia ziemi obawiają się mieszkańcy regionu Pacific, a najmniej mieszkańcy regiony East North Central.
