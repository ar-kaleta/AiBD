```python
Xcopy = X.copy()
```

```python
mapping = {"Not at all worried" : "Not worried", "Not so worried" : "Not worried"}
```

```python
Xcopy["In general, how worried are you about earthquakes?"] = \
Xcopy["In general, how worried are you about earthquakes?"].apply(lambda row: mapping.get(row, "Worried"))
Xcopy
```

```python
ct = pd.crosstab(Xcopy["US Region"], Xcopy["In general, how worried are you about earthquakes?"])
ct
```

```python
ct.plot.bar(figsize=(15,7))
plt.tight_layout()
```

```python
(ct["Not worried"]/ct["Worried"])
```

```python
(ct["Not worried"]/ct["Worried"]).plot.bar(figsize=(15,7))
plt.tight_layout()
plt.title('Stosunek osb nie martwicych si o trzsienie ziemi do osb obawiajcych si trzsienia ziemi')
```

```python
(ct["Not worried"]/ct["Worried"]).idxmax(), (ct["Not worried"]/ct["Worried"]).idxmin()
```

Stosunek osb ktre nie boj si trzsienia ziemi do stosunku osb ktre boj si trzsienia ziemi jest najwikszy w regionine East North Central, natomiast najwikszy w regionie Pacific. Mona zatem powiedzie, e najbardziej trzsienia ziemi obawiaj si mieszkacy regionu Pacific, a najmniej mieszkacy regiony East North Central.
